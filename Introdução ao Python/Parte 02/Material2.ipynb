## 🧩&nbsp;&nbsp;MODULARIZAÇÃO

---

A modularidade trata da reutilização de código, abstração de complexidade e facilitação de manutenibilidade. A construção de funções permitem o isolamento de blocos de código que executam tarefas específicas, facilitando a leitura, a manutenção e o teste. Ao encapsular a lógica em funções, podemos evitar a redundância, reduzir erros e garantir que as alterações feitas em um ponto do código sejam refletidas de maneira consistente em toda a aplicação. Além disso, funções facilitam a colaboração e a escalabilidade, pois diferentes desenvolvedores podem trabalhar em funções distintas sem interferir no trabalho uns dos outros.


### Funções pré-definidas
Existes centenas de funções que já estão pré-definidas, seja pela própria linguagem ou seja pelo uso de bibliotecas e pacotes adicionais.

````python
#Simple Rounding Functions
a=3.14
a=round(a)
print(a)
````

````python
#We can string functions together
import numpy as np
b = np.arange(15).reshape(3, 5)
print(b)
````

###Customização de funções


Para criarmos funções que poderão ser chamadas de acordo com a necessidade, usamos a palavra reserva ``` def ``` seguida do nome da função e os sinais de parênteses para indicar a presença ou não de parâmetros.


```python
def function_name(argument1, argument2,...,):
    """
    Description of function
    """
    
    # Function statements
    return(return_value)
```



````python
def quadrado(x):
  return x ** 2

def multiplicacao(a, b):
  return a * b
````

````python
# chamada de função com um parâmetro
quadrado(5)
````

````python
# chamada de função com dois parâmetros
multiplicacao(3, 2)
````

````python
# Composição de funções
quadrado(multiplicacao(3, 2))
````

### Parâmetros e retorno em funções

Assim como em outras linguagens de programação podemos construir diferentes tipos de funções em termos de passagem de parâmetros e existência ou não de instruções de retorno.

Podemos passar quantos parâmetros forem necessários para uma função:


````python
from datetime import datetime

# Declaração de função que NÃO TEM parâmetros e NÃO TEM retorno
def boas_vindas_fixo():
    print("Seja bem vindo(a)!")

# Declaração de função que TEM parâmetros e NÃO TEM retorno
def boas_vindas_dinamico(mensagem):
    print(mensagem)

# Declaração de função que NÃO TEM parâmetros e TEM RETORNO
def boas_vindas_com_horario():
    return "Seja bem vindo! Agora são " + datetime.now().strftime("%H:%M de %d/%m/%Y")

# Declaração de função que TEM parâmetros e TEM RETORNO
def boas_vindas_com_nome(nome):
    return "Seja bem vindo(a) " + nome + "!"
````

````python
# Chamada da função que NÃO TEM parâmetros e NÃO TEM retorno
boas_vindas_fixo()
````

````python
# Chamada da função que TEM parâmetros e NÃO TEM retorno
boas_vindas_dinamico("Olá! Seja muito bem vindo(a)! :)")
````

````python
# Declaração de função que NÃO TEM parâmetros e TEM RETORNO
mensagem_horario = boas_vindas_com_horario()
print("A mensagem retornada foi: " + mensagem_horario)
````

````python
# Chamada de função que TEM parâmetros e TEM RETORNO
mensagem_nome = boas_vindas_com_nome("Maria")
print("A mensagem retornada foi: " + mensagem_nome)
````

### Parâmetros com valores padrão

Podemos pré-definir valores padrão para parâmetros. Assim, se ao chamarmos uma função e não passarmos valores para estes parâmetros, automaticamente eles adotarão os valores padrões definidos em suas declarações.

Este primeiro exemplo apresenta quatro parâmetros e não define valores padrão. Logo ao realizar a chamada deve-se passar todos os valores para cada parâmetro ou indica-los como vazios passando "".

````python
def dados_usuario(nome, cpf, email, pais):
    print("Nome: " + nome)
    print("CPF: " + cpf)
    print("E-mail: " + email)
    print("País: " + pais)

dados_usuario("Pedro Santos", "123.456.789-10", "pedro@email.com", "Brasil")
````

Já no caso abaixo, note que os parâmetros de email e pais estão com valores pré-definidos na declaração da função. Logo, quando estes parâmetros não forem passados na chamada da função, eles vão automaticamente atribuir o valor que está como padrão na função.

Importante ressaltar que os parâmetros que tem valores padrão devem sempre estar definidos em sequência ao final da lista de parâmetros.

````python
def dados_usuario_padrao(nome, cpf, email = "Não informado", pais = "Brasil"):
    print("Nome: " + nome)
    print("CPF: " + cpf)
    print("E-mail: " + email)
    print("País: " + pais)

dados_usuario_padrao("Pedro Santos", "123.456.789-10")
````

### Funções com estruturas de dados

Nas funções, além de trabalhar com os valores únicos para cada parâmetro, podemos também utilizar estruturas de dados o que amplia muito as possibilidades! Sendo assim, podemos passar como parâmetros estruturas de dados como listas e dicionários.

````python
# Recebe uma lista como parâmetro e exibe todos os itens
def lista_compras(itens):
    for i in itens:
        print(i)
````

````python
# Define uma lista e passa esta estrutura como parâmetro na chamada da função
itens = ["Pão", "Café", "Manteiga", "Cereal"]
lista_compras(itens)
````

Da mesma forma, podemos também fazer com que o retorno de uma função seja uma estrutura de dados, tanto de listas, quanto de dicionários.

````python
import random

# Recebe uma lista e cria um dicionário que é retornado como parâmetro
def compras_quantidade(itens):
    completa = {}
    for i in itens:
        completa[i] = random.randint(2, 5)
    return completa
````

````python
# Chamada da função que retorna um dicionário armazenado em outra variável
dicionario_completa = compras_quantidade(itens)
print("Dicionário criado: ")
print(dicionario_completa)
````

Para apresentar as informações de chave e de valor do dicionário criado, podemos usar a função `items()` e fazer a iteração em cada item.

````python
for k, v in dicionario_completa.items():
    print("Comprar " + str(v) + " unidades de " + k)
````

### Retorno de múltiplos valores

````python
def calculos(a, b):
    adicao = a+b
    subtracao = a-b
    divisao = a/b
    multiplicacao = a*b
    return(adicao, subtracao, divisao, multiplicacao)
````

````python
a, s, d, m = calculos(10, 3)
multiplicacao = calculos(10, 3)[3]

total = a + s + d + m
print("Multiplicacao resultou em: " + str(multiplicacao))
print("Adição resultou em: " + str(a))
print("Valor total da soma dos resultados: " + str(total))
````

## TRATAMENTO DE EXCEÇÕES

### Lidando com exceções


É um evento que ocorre durante a execução de um programa quando há um desvio do fluxo normal das instruções programadas. Quando isso ocorre, cria-se então um um objeto que representa um erro. Para previnir e/ou expressar esses erros corretamente existe o tratamento de exceções. No Python existem dezenas de exceções que podem ser mapeadas, você pode encontrar uma lista completa delas em: https://docs.python.org/pt-br/3.12/library/exceptions.html#Exception

Assim, podemos criar o tratamento de exceção toda vez que verificarmos que há possibilidade que um erro aconteça.

````python
12/0
````

````python
try:
    12 / 0
except ZeroDivisionError:
    print('Não é possível dividir por zero!')
````

````python
try:
    lista = [1, 2, 3]
    print(lista[4])try
except:
    print('Um erro ocorreu!')
````

Por padrão, uma exceção gera um objeto que inclui informações sobre o problema que causou o erro.

````python
try:
    lista = [1, 2, 3]
    print(lista[4])
except Exception as erro:
    print("Um erro ocorreu: " + str(erro.args[0]))
````

### Hierarquia de exceções

As exceções devem ser organizadas conforme uma hieraquia, isso significa que podemos ter múltiplas exceções mapeadas.

````python
try:
   # 12/0
    lista = [1, 2, 3]
    print(lista[4])
except ZeroDivisionError:
    print("Não é possível dividir por zero ;)")
except Exception as erro:
    print("Um erro ocorreu!")
````

Estrutura Finally

A cláusula `finally`contém um bloco de código que sempre vai ser executado independentemente de haver alguma exceção no código ou não. Geralmente é usado para limpar alguns recursos de um programa, especialmente se estiveremos usando operações de E/S de arquivo.

````python
try:
    12/0
   # lista = [1, 2, 3]
   # print(lista[4])
except ZeroDivisionError:
    print("Não é possível dividir por zero ;)")
except Exception as erro:
    print("Um erro ocorreu!")
finally:
    print("Encerrando a aplicação...")
````
